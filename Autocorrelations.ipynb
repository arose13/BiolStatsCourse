```python
from statsImport import *
from statsmodels.tsa.stattools import adfuller
```

```python
x = np.linspace(0, 5*np.pi, 100)

data = pd.DataFrame(x, columns=['x'])
data['y'] = np.sin(x)

graph.title('data')
graph.plot(data.y)
graph.show()
```

```python
fx = data.y+data.x**2

graph.title('Difference $y = sin(x) + x^2$ until stationary')
graph.plot(fx, label='y')
graph.plot(fx.diff(), label='y\'')
graph.plot(fx.diff().diff(), label='y"')
graph.legend(loc=0)
graph.ylim([-5, 10])
graph.show()

print('0 order or y, p-value:', adfuller(fx)[1])
print('=========================\n')
print('1 order or y\', p-value:', adfuller(fx.diff().dropna())[1])
print('=========================\n')
print('2 order or y", p-value:', adfuller(fx.diff().diff().dropna())[1])
```

```python
graph.title('lags')
graph.plot(data.y.shift(1))
graph.plot(data.y.shift(100))
graph.plot(data.y)
graph.show()
```

```python
sns.jointplot(data.y, data.y.shift(5))
graph.show()

sns.jointplot(data.y, data.y.diff().diff())
graph.show()
```
